```python
import pandas as pd
import numpy as np
```

```python
# Load data
df = pd.read_csv("Airbnb_Open_Data.csv")
```

```python
df.columns
```

```python
#Clean Column Names (snake_case)
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# Know your data
df.head()
```

```python
pd.set_option('display.max_columns', 30)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# data types
df.dtypes
```

```python
# Remove Duplicates
df = df.drop_duplicates()
```

```python
df.head()
```

```python
# Handle Missing Values Logically
# Identify missing data
missing_summary = df.isnull().sum().sort_values(ascending=False)
print("Missing values per column:\n", missing_summary)
```

```python
# Fill numeric columns with 0 or median, string columns with mode or 'Unknown' only if logical
fill_zero_cols = ['minimum_nights', 'number_of_reviews', 'review_rate_number', 'calculated_host_listings_count', 'availability_365']
for col in fill_zero_cols:
    df[col] = df[col].fillna(0).astype(int)
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Clean and Convert Price & Fees Columns
for money_col in ['price', 'service_fee']:
    df[money_col] = (
        df[money_col]
        .astype(str)
        .replace('[\$,]', '', regex=True)
        .replace('nan', 0)
        .astype(float)
    )
```

```python
# Parse Dates & Handle Date Features
if 'last_review' in df.columns:
    df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
```

```python
df.head()
```

```python
#Standardize Categorical Columns (lowercase, no spaces)
cat_cols = ['host_identity_verified', 'cancellation_policy', 'room_type', 'neighbourhood']
for col in cat_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).str.lower().str.replace(' ', '_')
```

```python
#Fix Construction Year or Other Int Columns
if 'construction_year' in df.columns:
    df['construction_year'] = (
        df['construction_year']
        .fillna('2012')
        .astype(str)
        .str.extract(r'(\d{4})')
        .astype(int)
    )
```

```python
#Drop Unnecessary Columns
drop_cols = ['house_rules', 'country', 'country_code', 'name', 'license', 'reviews_per_month']
for col in drop_cols:
    if col in df.columns:
        df.drop(columns=[col], inplace=True)
```

```python
# Final Clean
df = df.dropna()  # Only if needed for downstream tasks
```

```python
df.head()
```

```python
#Save Cleaned Data
df.to_csv("airbnb_cleaned.csv", index=False)
```
